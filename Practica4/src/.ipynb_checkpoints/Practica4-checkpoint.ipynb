```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
from PIL import Image
import math
import operator as op
```

```python
#Leeemos la imagen
plt.rcParams['image.cmap'] = 'gray'
fourier = io.imread('images/fourier_bw512.bmp') #Imagen de Fourier

#Funcion para plotear de volon pimpon
def plot(objeto):
    plt.imshow(objeto)
```

```python
#Ejercicio 1

Tf = np.fft.fft2(fourier) #Discreta
sTf= np.fft.fftshift(Tf)

def magnitud(img):
    parte_realTf= abs(img) #Magnitud
    return parte_realTf
def fase(img):
    fase_Tf= np.angle(img) #Fase
    return fase_Tf
def amplitud(img):
    amplitud_Tf= np.log(1+img)
    return amplitud_Tf

def i_amp(img):
    IAmplitud = np.fft.ifft2(magnitud(img))
    IAmplitud = abs(IAmplitud)
    return IAmplitud

def i_fase(img):    
    IFase = np.fft.ifft2(1*img)
    IFase = abs(IFase)
    return IFase
```

```python
magF = magnitud(Tf)
smagF = magnitud(sTf)
#-------
faseF = fase(Tf)
sfaseF = fase(sTf)
#-------
ampF = amplitud(Tf)
sampF = amplitud(sTf)
#-------
iAmpF = i_amp(Tf)
siAmpF = i_amp(sTf)
#-------
iFasF = i_fase(Tf)
siFasF = i_fase(sTf)
```

```python
# Ejercicio 2
def butterworth(img,Dx,n,filtro):
    c,r = img.shape
    u = np.arange(r)
    v = np.arange(c)
    u,v = np.meshgrid(u,v) # matriz de u y v
    forma = np.sqrt((u- (r/2))**2+ (v- (c/2))**2)
    d = (1.0 + (forma/Dx)**(2*n))
    tras_Fourier = np.fft.fftshift(np.fft.fft2(img)) #Fourier
    con_filtro = op.mul(tras_Fourier,filtro(d))
    Icon_filtro = np.fft.ifft2(np.fft.ifftshift(con_filtro)).real #Fourier a la inversa
    return Icon_filtro

# Filtro paso bajas
def paso_bajas(d):
    filtro = 1.0/d
    return filtro

# Filtro paso altas
def paso_altas(d):
    filtro = 1-(1.0/d)
    return filtro
    
```

```python
btw_bajas = butterworth(fourier,145,2,paso_bajas)
btw_altas = butterworth(fourier,145,2,paso_altas)
```

```python
plot(btw_bajas)
```

```python
plot(btw_altas)
```
